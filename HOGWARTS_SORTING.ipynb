```python
house_mapping = {
    'Gryffindor': 0,
    'Slytherin': 1,
    'Hufflepuff': 2,
    'Ravenclaw': 3
}
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
x = df[['BLOOD STATUS_NUM', 'Species_NUM', 'GENDER_NUM', 'HAIR COLOR_NUM', 'BOGGART_NUM', 'PATRONUS_NUM', 'Nationality_NUM', 'ALIVE?_NUM']]
y = df['HOUSE_NUM']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(x_train, y_train)
```

```python
y_pred = rf_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"Random Forest Classification Model Accuracy: {accuracy*100}%")
# with
```

```python
from sklearn.metrics import precision_score, accuracy_score, classification_report, confusion_matrix, recall_score, f1_score

classification_rep = classification_report(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
confusion = confusion_matrix(y_test, y_pred)

print("Classification Report:\n", classification_rep)
print("Precision:", precision)
print("Recall (Sensitivity):", recall)
print("F1-Score:", f1)
print("Confusion Matrix:\n", confusion)
```

```python
from sklearn.metrics import confusion_matrix

confusion = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=house_mapping.keys(), yticklabels=house_mapping.keys())
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
